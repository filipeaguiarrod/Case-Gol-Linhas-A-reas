Possível observar queda expressiva  entre meses 02-03 de 2020 consequência da pandemia.

O interessante seria um possível análise com série temporal para estes dados.

### Payload(Capacidade) x Carga paga

```python
#Coluna com relação Payload x Carga Paga:

df_s['carga_paga_payload'] = df_s['kg_carga_paga']/df_s['kg_payload']

df_s['carga_paga_payload'].describe()
```

```python
# Corrigindo valores sem sentido que podem prejudicar nas "plotagens"


df_s.loc[df_s['carga_paga_payload']>1,'carga_paga_payload']=np.nan
df_s.loc[df_s['carga_paga_payload']<=0,'carga_paga_payload']=np.nan
```

```python
df_s['carga_paga_payload'].describe()
```

```python
#Observando distribuição da relação carga paga/payload disponivel


plt.figure(figsize=(12.5,12.5))
sns.displot(df_s, x="carga_paga_payload")
plt.xlim(0,0.2)
```

```python
sns.boxplot(df_s.loc[df_s['sg_empresa_icao'] == 'GLO','carga_paga_payload'])
```

```python
plt.figure(figsize=(12.5,12.5))
plt.title('GOL Linhas Aéreas')
sns.scatterplot(x=df_s.loc[df_s['sg_empresa_icao']=='GLO','kg_payload'] ,y = df_s.loc[df_s['sg_empresa_icao']=='GLO','kg_carga_paga'])
plt.xticks(rotation=90);
```

Possível observar duas linhas, de possíveis capacidade de kg_payload, duas possíves aeronaves. Para uma mesma capacida o valor que foi pago vária de
0 a 4000 e 6000 kg, muito longe da capacidade máxima, pensando que no mundo ideal 100% do payload fosse pago.
