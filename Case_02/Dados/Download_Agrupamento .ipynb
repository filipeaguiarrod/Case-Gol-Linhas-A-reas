### Estratégia de Extração dos Dados

Observada a quantidade de dados que foi disponibiliza-da foi necessário criar uma estratégia para fazer o download dos dados de uma forma automatizada visto que são aproximadamente: 502 fontes para download. 

A estratégia de resolução para o download múltiplo veio entre outros links, principalmente da fonte abaixo:

https://towardsdatascience.com/how-to-web-scrape-with-python-in-4-minutes-bc49186a8460

```python
import numpy as np
import pandas as pd
import seaborn as sns
import requests
import urllib.request
import time
from bs4 import BeautifulSoup
import os 
```

```python
# !pip install requests
```

```python
url = 'https://www.anac.gov.br/assuntos/setor-regulado/empresas/envio-de-informacoes/microdados'
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
soup.findAll('a')
```

```python
for i in range(84,590):
         
    one_a_tag = soup.findAll('a')[i] # percorre a linha com o index de i 
    link = one_a_tag['href'] # retorna o link 
    download_url = link # Alias 
    time.sleep(1) # Tempo de resposta para não sobrecarregar o servidor 
    r = requests.get(link, allow_redirects=True) # Acesso 
    urllib.request.urlretrieve(link,download_url[len(download_url)-17:]) #Download no diretório de trabalho com nome vindo da própria url.
```

Todos os arquivos foram baixados em zip, e mesmo compactados tem um tamanho total de 2 Gb aproximadamente, que deverá ser expressivamente aumentado quando descompactados
